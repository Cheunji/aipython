**키 값에 대한 몸무게 예측**

```python
import numpy as np
from sklearn import linear_model # scikit-learn 모듈을 가져옴
import matplotlib.pyplot as plt

regr = linear_model.LinearRegression() # 선형회귀 모델 생성

X = [[164],[179],[162],[170]]
y = [53, 63, 55, 59]
regr.fit(X,y)

plt.scatter(X,y)                          # 시각화
y_pred = regr.predict(X)
plt.plot(X, y_pred, color = 'blue', linewidth = 3)
plt.show()

# score = regr.score(X, y)
# print("The score of this line for the data: ", score)

input_data = [[180],[185]]
result = regr.predict(input_data)
print(result)
```

**학생의 과목수와 공부시간으로 시험 점수 예측**

```python
import pandas as pd
std_df = pd.read_csv("Student_Marks.csv")

from sklearn.model_selection import train_test_split
x = std_df.drop('Marks', axis=1) # std_df['number_course', 'time_study']
y = std_df['Marks']
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.2, random_state=42)

from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()

lr_model.fit(X_train, Y_train)

lr_pred = lr_model. predict(X_test)

# 모델 평가

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

print("r2_score: ", r2_score(lr_pred, Y_test))
print("mean_squared_error: ", mean_squared_error(lr_pred, Y_test))
```

**결정계수(-1~1)**  ... r2_score
1에 가까울수록 좋은 수치

**평균 제곱 오차** ... mean_squared_error
예측값과 실제값 사이의 얼마나 차이나는지 수치화(값이 작을수록 예측 성능 good)
