```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**타이타닉 데이터 읽어와서 확인하기**

```python
# 1. 데이터 파일 읽어오기

titanic_data = pd.read_csv('titanic.csv')
titanic_data
```

```python
# 2. 데이터의 기본 정보 출력

titanic_data.info()
```

```python
# 3. 데이터의 기초 통계량 출력

titanic_data.describe()
```

```python
# 4. 요금(fare) 기준 내림차순 정렬

titanic_data2 = titanic_data.sort_values(by=['fare'], ascending=False)
titanic_data2
```

```python
# 5. 생존자별 인원수 확인

titanic_data['survived'].value_counts()
```

**타이타닉 데이터 전처리**

```python
# 1. 결측치 확인하기

titanic_data.isna().sum()
```

```python
# 2. 컬럼 삭제하기
# 결측치가 많은 'Deck' 컬럼 삭제

titanic_data.drop('deck', axis=1, inplace=True)
```

```python
# 3. 최빈값 대체하기
# 결측치가 작은 'Embarked' 컬럼에 결측치를 최빈값으로 대체

titanic_data.groupby('pclass')
titanic_data.fillna({'embarked':'S'}, inplace=True)
```

```python
# 3-1. 탑승 항구 최빈값 확인 : S

titanic_data['embarked'].value_counts()
```

```python
# 4. 평균값 대체하기
# 나이의 결측치를 평균 데이터로 대체

titanic_data.fillna({'age':titanic_data['age'].mean()}, inplace=True)
```

```python
titanic_data.isna().sum()
```

**타이타닉 데이터 그룹핑**

```python
# 1. 항구(embarked)별 pclass의 평균값 구하기

titanic_data.groupby('embarked')['pclass'].mean()
```

```python
# 2. pclass, sex 컬럼 기준으로 평균 구하기

titanic_data.groupby(['pclass', 'sex']).mean()
```

**타이타닉 데이터 시각화**

```python
# 1. 막대 그래프 그리기
# 좌석 등급별 생존자 확인(seaborn 활용)

sns.countplot(x='pclass', hue='survived', data=titanic_data)
```

```python
# 2. 히트맵 그래프 그리기
# 푸른색, 숫자 표현
titanic_data3 = titanic_data.corr()
sns.heatmap(data=titanic_data3, annot=True, fmt='.2f', cbar=False, cmap='Blues')
```
